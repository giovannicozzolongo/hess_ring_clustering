# Unsupervised Clustering of FlashCam Images

This notebook applies unsupervised learning to FlashCam telescope images from H.E.S.S. The goal is to see if a neural network can automatically group similar air shower images without being told what to look for.

**Background**: When cosmic rays hit the atmosphere, they create cascades of particles that emit Cherenkov light. The FlashCam camera captures images of these events. Different types of events (gamma-ray showers, hadron showers, muons) produce characteristic image patterns. Rather than hand-coding rules to identify these patterns, we'll use unsupervised learning to discover natural groupings in the data.

**The approach**:
1. Load FlashCam images that were preprocessed and saved to HDF5 format
2. Use a pre-trained convolutional neural network (VGG19) to extract features from each image
3. Apply clustering algorithms (K-means and Gaussian Mixture Models) to group similar images
4. Visualize and interpret the clusters

**Note**: We don't have ground truth labels for our data, so this is purely exploratory. We're asking: "what patterns does the network find?"

## Import libraries

We'll use:
- **numpy/tables**: for loading HDF5 data
- **matplotlib**: for plotting
- **keras**: to load pre-trained CNNs
- **sklearn**: for clustering algorithms and dimensionality reduction

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import tables
import os
import random
from tqdm import tqdm

from tensorflow import keras
from keras.applications import VGG19
from keras.applications.vgg19 import preprocess_input

from sklearn.cluster import KMeans
from sklearn.mixture import GaussianMixture
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split

# Set random seeds for reproducibility
np.random.seed(42)
random.seed(42)

print("✓ Libraries loaded successfully")
```

## Load FlashCam data

The data should be in HDF5 files containing:
- **images**: 2D arrays (square grid representation of camera images)
- **event_nr**: event IDs for tracking
- **dst_nr**: DST file numbers

We'll load the images into `x_train` and `x_test` arrays, just like in the cats/dogs example. However, we won't have `y_train` or `y_test` since we don't have labels.

```python
def load_flashcam_data(data_dir, max_images=None, test_split=0.2):
    """
    Load FlashCam images from HDF5 files.
    
    Parameters:
    -----------
    data_dir : str
        Directory containing HDF5 files
    max_images : int or None
        Maximum number of images to load (None = all)
    test_split : float
        Fraction of data to use for testing (0.2 = 20%)
    
    Returns:
    --------
    x_train, x_test : numpy arrays
        Training and test images
    event_info_train, event_info_test : numpy arrays
        Event numbers for tracking
    """
    
    all_images = []
    all_event_nrs = []
    
    # Find all HDF5 files in the directory
    h5_files = sorted([f for f in os.listdir(data_dir) if f.endswith('.h5')])
    print(f"Found {len(h5_files)} HDF5 files")
    
    # Load images from each file
    for h5_file in h5_files:
        filepath = os.path.join(data_dir, h5_file)
        print(f"Loading {h5_file}...", end=' ')
        
        with tables.open_file(filepath, 'r') as f:
            images = f.root.images[:]
            event_nrs = f.root.event_nr[:]
            
            all_images.append(images)
            all_event_nrs.append(event_nrs)
            
            print(f"{len(images)} images")
        
        if max_images and len(np.concatenate(all_images)) >= max_images:
            break
    
    # Concatenate all data
    all_images = np.concatenate(all_images)
    all_event_nrs = np.concatenate(all_event_nrs)
    
    # Limit to max_images if specified
    if max_images and len(all_images) > max_images:
        indices = np.random.choice(len(all_images), max_images, replace=False)
        all_images = all_images[indices]
        all_event_nrs = all_event_nrs[indices]
    
    print(f"\nTotal images loaded: {len(all_images)}")
    print(f"Image shape: {all_images[0].shape}")
    
    # Split into train/test
    x_train, x_test, evt_train, evt_test = train_test_split(
        all_images, all_event_nrs, test_size=test_split, random_state=42
    )
    
    print(f"Training set: {len(x_train)} images")
    print(f"Test set: {len(x_test)} images")
    
    return x_train, x_test, evt_train, evt_test
```

```python
# Specify the directory containing your HDF5 files
DATA_DIR = '../data/flashcam'

# Load data - start with a smaller subset for testing
# Remove max_images limit to use all available data
x_train, x_test, evt_train, evt_test = load_flashcam_data(
    DATA_DIR, 
    max_images=10000  # Start with 10k images, increase as needed
)
```

```python
# Replace NaN values with 0 (pixels outside the camera)
print("=" * 60)
print("CLEANING NaN VALUES")
print("=" * 60)

print(f"\nBefore cleaning:")
print(f"  Training set NaN count: {np.isnan(x_train).sum()} / {x_train.size} ({np.isnan(x_train).sum()/x_train.size*100:.2f}%)")
print(f"  Test set NaN count: {np.isnan(x_test).sum()} / {x_test.size} ({np.isnan(x_test).sum()/x_test.size*100:.2f}%)")

# Replace NaN with 0 (these are pixels outside the camera circle)
x_train = np.nan_to_num(x_train, nan=0.0)
x_test = np.nan_to_num(x_test, nan=0.0)

print(f"\nAfter cleaning:")
print(f"  Training set NaN count: {np.isnan(x_train).sum()}")
print(f"  Test set NaN count: {np.isnan(x_test).sum()}")

print(f"\nTraining set statistics:")
print(f"  Min: {x_train.min():.6f}, Max: {x_train.max():.6f}")
print(f"  Mean: {x_train.mean():.6f}, Std: {x_train.std():.6f}")

print("\n✓ NaN values replaced with 0")
print("=" * 60)
```

```python
# Filter out problematic images before processing
print("=" * 60)
print("FILTERING PROBLEMATIC IMAGES")
print("=" * 60)

def is_image_valid(img, min_std=1e-6, min_range=1e-6):
    """
    Check if an image is valid for processing.
    Returns (is_valid, reason)
    """
    # Check for empty/constant images
    img_range = img.max() - img.min()
    if img_range < min_range:
        return False, "constant/empty"
    
    # Check for very low variance
    img_std = np.std(img)
    if img_std < min_std:
        return False, "low variance"
    
    # Check for NaN or Inf
    if np.isnan(img).any():
        return False, "contains NaN"
    if np.isinf(img).any():
        return False, "contains Inf"
    
    return True, "valid"

# Analyze training set
print(f"\nAnalyzing training set ({len(x_train)} images)...")
valid_mask_train = []
problem_reasons_train = {}
problematic_indices_train = []

for i, img in enumerate(x_train):
    is_valid, reason = is_image_valid(img)
    valid_mask_train.append(is_valid)
    
    if not is_valid:
        problem_reasons_train[reason] = problem_reasons_train.get(reason, 0) + 1
        if len(problematic_indices_train) < 5:  # Keep first 5 for visualization
            problematic_indices_train.append((i, reason))

valid_mask_train = np.array(valid_mask_train)
n_problematic_train = (~valid_mask_train).sum()

print(f"✓ Valid images: {valid_mask_train.sum()}")
print(f"✗ Problematic images: {n_problematic_train}")
if problem_reasons_train:
    print("\nReasons for rejection:")
    for reason, count in problem_reasons_train.items():
        print(f"  - {reason}: {count}")

# Analyze test set
print(f"\nAnalyzing test set ({len(x_test)} images)...")
valid_mask_test = []
problem_reasons_test = {}
problematic_indices_test = []

for i, img in enumerate(x_test):
    is_valid, reason = is_image_valid(img)
    valid_mask_test.append(is_valid)
    
    if not is_valid:
        problem_reasons_test[reason] = problem_reasons_test.get(reason, 0) + 1
        if len(problematic_indices_test) < 5:
            problematic_indices_test.append((i, reason))

valid_mask_test = np.array(valid_mask_test)
n_problematic_test = (~valid_mask_test).sum()

print(f"✓ Valid images: {valid_mask_test.sum()}")
print(f"✗ Problematic images: {n_problematic_test}")
if problem_reasons_test:
    print("\nReasons for rejection:")
    for reason, count in problem_reasons_test.items():
        print(f"  - {reason}: {count}")

# Show examples of problematic images
if problematic_indices_train:
    print("\n" + "=" * 60)
    print("EXAMPLES OF PROBLEMATIC IMAGES (TRAINING SET)")
    print("=" * 60)
    
    n_examples = min(len(problematic_indices_train), 5)
    fig, axes = plt.subplots(1, n_examples, figsize=(15, 3))
    if n_examples == 1:
        axes = [axes]
    
    for i, (idx, reason) in enumerate(problematic_indices_train[:n_examples]):
        img = x_train[idx]
        axes[i].imshow(img, cmap='viridis')
        axes[i].set_title(f'Index {idx}\n{reason}\nmin={img.min():.2e}\nmax={img.max():.2e}')
        axes[i].axis('off')
    
    plt.tight_layout()
    plt.show()
    
    # Show statistics
    print("\nStatistics of problematic images:")
    for idx, reason in problematic_indices_train[:n_examples]:
        img = x_train[idx]
        print(f"  Index {idx} ({reason}): min={img.min():.3e}, max={img.max():.3e}, "
              f"mean={img.mean():.3e}, std={img.std():.3e}")

# Filter the datasets
print("\n" + "=" * 60)
print("APPLYING FILTER")
print("=" * 60)

x_train_filtered = x_train[valid_mask_train]
evt_train_filtered = evt_train[valid_mask_train]

x_test_filtered = x_test[valid_mask_test]
evt_test_filtered = evt_test[valid_mask_test]

print(f"\nOriginal training set: {len(x_train)} images")
print(f"Filtered training set: {len(x_train_filtered)} images ({len(x_train_filtered)/len(x_train)*100:.1f}%)")

print(f"\nOriginal test set: {len(x_test)} images")
print(f"Filtered test set: {len(x_test_filtered)} images ({len(x_test_filtered)/len(x_test)*100:.1f}%)")

# Replace original arrays with filtered ones
x_train = x_train_filtered
evt_train = evt_train_filtered
x_test = x_test_filtered
evt_test = evt_test_filtered

print("\n✓ Datasets updated with filtered images")
print("=" * 60)
```

## Prepare images for VGG19

VGG19 was trained on color images (3 channels: RGB) with shape (224, 224, 3). Our FlashCam images are grayscale. We need to:
1. Resize to 224×224 if needed
2. Convert to 3 channels (just replicate the grayscale across RGB)
3. Normalize pixel values

```python
def prepare_images_for_vgg(images):
    """
    Prepare FlashCam images for VGG19:
    - Resize to 224x224 if needed
    - Convert grayscale to 3-channel RGB
    - Apply VGG preprocessing
    """
    from skimage.transform import resize
    
    prepared = []
    
    print(f"Preparing {len(images)} images for VGG19...")
    
    for img in tqdm(images):
        # Resize to 224x224 if needed
        if img.shape != (224, 224):
            img = resize(img, (224, 224), anti_aliasing=True)
        
        # Normalize to [0, 255] range
        img = (img - img.min()) / (img.max() - img.min() + 1e-8)
        img = img * 255.0
        
        # Convert to 3 channels (RGB)
        img_rgb = np.stack([img, img, img], axis=-1)
        
        prepared.append(img_rgb)
    
    prepared = np.array(prepared, dtype=np.float32)
    
    # Apply VGG-specific preprocessing
    prepared = preprocess_input(prepared)
    
    return prepared
```

```python
# Prepare training and test sets
x_train_prepared = prepare_images_for_vgg(x_train)
x_test_prepared = prepare_images_for_vgg(x_test)

print(f"\nPrepared training data shape: {x_train_prepared.shape}")
print(f"Prepared test data shape: {x_test_prepared.shape}")
```

## Load pre-trained VGG19 model

VGG19 is a convolutional neural network trained on millions of everyday images (ImageNet dataset). Even though it wasn't trained on telescope images, its learned features (edges, textures, shapes) are general enough to be useful for our task.

We'll use it as a **feature extractor**: instead of using the final classification layer, we take the output from deeper in the network. This gives us a compact numerical representation of each image that captures its important visual features.

```python
# Load VGG19 without the top classification layers
# include_top=False means we only get the convolutional base
# pooling='avg' adds global average pooling at the end

print("Loading VGG19 model...")
vgg19_model = VGG19(
    weights='imagenet',
    include_top=False,
    pooling='avg',
    input_shape=(224, 224, 3)
)

print("✓ VGG19 loaded successfully")
print(f"Output shape: {vgg19_model.output_shape}")
print(f"\nThis means each image will be represented by {vgg19_model.output_shape[1]} numbers")
```

## Extract features from images

Now we pass our FlashCam images through VGG19 to get feature vectors. This will take a few minutes depending on how many images you loaded.

```python
def extract_features(model, images, batch_size=32):
    """
    Extract features from images using the pre-trained model.
    """
    print(f"Extracting features from {len(images)} images...")
    features = model.predict(images, batch_size=batch_size, verbose=1)
    print(f"✓ Feature extraction complete. Shape: {features.shape}")
    return features
```

```python
# Extract features
features_train = extract_features(vgg19_model, x_train_prepared)
features_test = extract_features(vgg19_model, x_test_prepared)
```

## Dimensionality reduction with PCA

The feature vectors from VGG19 are 512-dimensional, which can be computationally expensive for clustering. We'll use Principal Component Analysis (PCA) to reduce this to a smaller number of dimensions while keeping most of the important information.

PCA finds the directions in the data with the most variation. Think of it like finding the best angle to view a 3D object in 2D - you want to keep as much structure as possible.

```python
# Try different numbers of PCA components
n_components = 50  # You can experiment with this

print(f"Applying PCA to reduce from {features_train.shape[1]} to {n_components} dimensions...")

pca = PCA(n_components=n_components, random_state=42)
features_train_pca = pca.fit_transform(features_train)
features_test_pca = pca.transform(features_test)

print(f"✓ PCA complete")
print(f"Variance explained: {pca.explained_variance_ratio_.sum():.2%}")
print(f"This means we kept {pca.explained_variance_ratio_.sum():.2%} of the original information")
```

```python
# Plot explained variance
plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('Number of Components')
plt.ylabel('Cumulative Explained Variance')
plt.title('How much information do we keep?')
plt.grid(True, alpha=0.3)

plt.subplot(1, 2, 2)
plt.bar(range(1, min(21, n_components+1)), pca.explained_variance_ratio_[:20])
plt.xlabel('Component Number')
plt.ylabel('Explained Variance Ratio')
plt.title('First 20 Components')
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

## K-Means Clustering

K-means is a simple clustering algorithm. It tries to partition the data into K groups by:
1. Randomly placing K cluster centers
2. Assigning each point to its nearest center
3. Moving each center to the average of its assigned points
4. Repeating steps 2-3 until convergence

The challenge: we need to choose K (number of clusters) ourselves. Let's try a few different values.

```python
# Try different numbers of clusters
k_values = [3, 5, 7, 10]

kmeans_models = {}
kmeans_predictions = {}

for k in k_values:
    print(f"\nRunning K-means with {k} clusters...")
    
    kmeans = KMeans(
        n_clusters=k,
        random_state=42,
        n_init=10,
        max_iter=300
    )
    
    # Fit on training data
    labels_train = kmeans.fit_predict(features_train_pca)
    labels_test = kmeans.predict(features_test_pca)
    
    kmeans_models[k] = kmeans
    kmeans_predictions[k] = {
        'train': labels_train,
        'test': labels_test
    }
    
    # Show cluster sizes
    unique, counts = np.unique(labels_train, return_counts=True)
    print(f"Cluster sizes (training): {dict(zip(unique, counts))}")
    print(f"Inertia (sum of squared distances to centers): {kmeans.inertia_:.2f}")

print("\n✓ K-means clustering complete")
```

## Visualize K-means clusters

Let's look at example images from each cluster to see if the algorithm found meaningful patterns.

```python
def show_cluster_examples(images, labels, k, n_examples=6):
    """
    Display example images from each cluster.
    """
    fig, axes = plt.subplots(k, n_examples, figsize=(15, 2.5*k))
    
    if k == 1:
        axes = axes.reshape(1, -1)
    
    for cluster_id in range(k):
        # Find images in this cluster
        cluster_indices = np.where(labels == cluster_id)[0]
        
        # Randomly select examples
        if len(cluster_indices) >= n_examples:
            selected = np.random.choice(cluster_indices, n_examples, replace=False)
        else:
            selected = cluster_indices
        
        for i, idx in enumerate(selected):
            if i < n_examples:
                axes[cluster_id, i].imshow(images[idx], cmap='viridis')
                axes[cluster_id, i].axis('off')
        
        # Hide unused subplots
        for i in range(len(selected), n_examples):
            axes[cluster_id, i].axis('off')
        
        # Label the row
        axes[cluster_id, 0].set_ylabel(
            f'Cluster {cluster_id}\n({len(cluster_indices)} images)',
            fontsize=12, rotation=0, labelpad=60, va='center'
        )
    
    plt.suptitle(f'K-means with {k} clusters - Random examples from each cluster', 
                 fontsize=14, y=1.00)
    plt.tight_layout()
    plt.show()

# Show examples for each K value
for k in k_values:
    print(f"\n=== K = {k} ===")
    show_cluster_examples(x_train, kmeans_predictions[k]['train'], k)
```

## Gaussian Mixture Model (GMM)

GMM is a more sophisticated clustering method than K-means. Instead of hard assignments ("this point belongs to cluster 3"), it gives probabilities ("this point is 70% cluster 3, 25% cluster 1, 5% cluster 2").

GMM assumes each cluster is a Gaussian (bell curve) distribution. This is more flexible than K-means and can handle clusters of different shapes and sizes.

```python
# Try GMM with different numbers of components
gmm_models = {}
gmm_predictions = {}

for k in k_values:
    print(f"\nRunning GMM with {k} components...")
    
    gmm = GaussianMixture(
        n_components=k,
        covariance_type='full',
        random_state=42,
        max_iter=100
    )
    
    # Fit on training data
    labels_train = gmm.fit_predict(features_train_pca)
    labels_test = gmm.predict(features_test_pca)
    
    gmm_models[k] = gmm
    gmm_predictions[k] = {
        'train': labels_train,
        'test': labels_test
    }
    
    # Show cluster sizes
    unique, counts = np.unique(labels_train, return_counts=True)
    print(f"Cluster sizes (training): {dict(zip(unique, counts))}")
    print(f"Log-likelihood: {gmm.score(features_train_pca) * len(features_train_pca):.2f}")

print("\n✓ GMM clustering complete")
```

## Visualize GMM clusters

```python
# Show examples for each K value
for k in k_values:
    print(f"\n=== K = {k} ===")
    show_cluster_examples(x_train, gmm_predictions[k]['train'], k)
```

## Visualize in 2D using PCA

To get an intuition for how well the clustering worked, let's project the data down to 2D and color points by their cluster assignment.

```python
# Further reduce to 2D for visualization
pca_2d = PCA(n_components=2, random_state=42)
features_2d = pca_2d.fit_transform(features_train_pca)

print(f"2D projection explains {pca_2d.explained_variance_ratio_.sum():.2%} of the (already reduced) variance")
```

```python
def plot_clusters_2d(features_2d, labels, k, title):
    """
    Plot clusters in 2D PCA space.
    """
    plt.figure(figsize=(10, 8))
    
    scatter = plt.scatter(
        features_2d[:, 0], 
        features_2d[:, 1],
        c=labels,
        cmap='tab10',
        alpha=0.5,
        s=10
    )
    
    plt.colorbar(scatter, label='Cluster ID')
    plt.xlabel('First Principal Component')
    plt.ylabel('Second Principal Component')
    plt.title(title)
    plt.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.show()

# Plot for K-means
k = 5  # Choose one value to visualize
plot_clusters_2d(
    features_2d, 
    kmeans_predictions[k]['train'], 
    k,
    f'K-means clusters (K={k}) projected to 2D'
)

# Plot for GMM
plot_clusters_2d(
    features_2d, 
    gmm_predictions[k]['train'], 
    k,
    f'GMM clusters (K={k}) projected to 2D'
)
```

## Compare cluster assignments

How much do K-means and GMM agree? Let's check how often they put the same images in the same cluster.

```python
from sklearn.metrics import adjusted_rand_score, normalized_mutual_info_score

print("Agreement between K-means and GMM:\n")
print("K    ARI    NMI")
print("-" * 20)

for k in k_values:
    kmeans_labels = kmeans_predictions[k]['train']
    gmm_labels = gmm_predictions[k]['train']
    
    ari = adjusted_rand_score(kmeans_labels, gmm_labels)
    nmi = normalized_mutual_info_score(kmeans_labels, gmm_labels)
    
    print(f"{k:2d}   {ari:.3f}  {nmi:.3f}")

print("\nARI = Adjusted Rand Index (1.0 = perfect agreement)")
print("NMI = Normalized Mutual Information (1.0 = perfect agreement)")
```

## Analyze cluster characteristics

Let's compute some basic statistics for each cluster to understand what distinguishes them.

```python
def analyze_cluster_statistics(images, labels, k):
    """
    Compute basic statistics for each cluster.
    """
    print(f"\nCluster statistics (K={k}):\n")
    print(f"{'Cluster':<10} {'Size':<10} {'Mean intensity':<15} {'Std intensity':<15}")
    print("-" * 60)
    
    for cluster_id in range(k):
        cluster_images = images[labels == cluster_id]
        
        mean_intensity = np.mean(cluster_images)
        std_intensity = np.std(cluster_images)
        
        print(f"{cluster_id:<10} {len(cluster_images):<10} {mean_intensity:<15.4f} {std_intensity:<15.4f}")

# Analyze K-means clusters
k = 5
print("=" * 60)
print("K-MEANS")
analyze_cluster_statistics(x_train, kmeans_predictions[k]['train'], k)

print("\n" + "=" * 60)
print("GMM")
analyze_cluster_statistics(x_train, gmm_predictions[k]['train'], k)
```

## Optional: t-SNE visualization

t-SNE (t-Distributed Stochastic Neighbor Embedding) is another dimensionality reduction technique that often produces better visualizations than PCA. It's slower but preserves local structure better.

```python
from sklearn.manifold import TSNE

# Use a subset of data for t-SNE (it's slow on large datasets)
n_samples = min(2000, len(features_train_pca))
indices = np.random.choice(len(features_train_pca), n_samples, replace=False)

print(f"Running t-SNE on {n_samples} samples (this may take a few minutes)...")

tsne = TSNE(n_components=2, random_state=42, perplexity=30)
features_tsne = tsne.fit_transform(features_train_pca[indices])

print("✓ t-SNE complete")
```

```python
# Plot t-SNE visualization
k = 5
labels_subset = kmeans_predictions[k]['train'][indices]

plt.figure(figsize=(12, 10))
scatter = plt.scatter(
    features_tsne[:, 0],
    features_tsne[:, 1],
    c=labels_subset,
    cmap='tab10',
    alpha=0.6,
    s=20
)
plt.colorbar(scatter, label='Cluster ID')
plt.xlabel('t-SNE dimension 1')
plt.ylabel('t-SNE dimension 2')
plt.title(f't-SNE visualization of clusters (K={k})')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```
